```python
import pandas as pd
```

### Tags

```python
tags = pd.read_csv("Tags.csv", engine='python')

tagsAndParents = tags[["Name", "Id"]]
tagsAndParents = tagsAndParents.rename(columns={"Id": "ParentId", "Name": "ParentTagName"})

tagsAndParents = pd.merge(tags, tagsAndParents, left_on="ParentTagId", right_on="ParentId")
tagsAndParents = tagsAndParents.drop(['ParentTagId', 'ParentId'], axis=1)

tagsAndParents.head()
```

### Datasets

```python
datasets = pd.read_csv("Datasets.csv", engine='python')
datasets.head()
```

```python
datasetTags = pd.read_csv("DatasetTags.csv", engine='python')
datasetTags.head()
```

```python
datasetsAndTags = pd.merge(datasetTags[["DatasetId", "TagId"]], tagsAndParents, left_on="TagId", right_on="Id")
datasetsAndTags = datasetsAndTags.drop(['Id', 'TagId'], axis=1)

datasetsAndTags = pd.merge(datasets, datasetsAndTags, left_on="Id", right_on="DatasetId")
datasetsAndTags = datasetsAndTags.drop(['DatasetId'], axis=1)

datasetsAndTags.head()
```

```python
datasetTagAndValue = datasetsAndTags.groupby(['Name', 'ParentTagName'])['Name'].count()
datasetTagAndValue
```

```python
datasetTagAndValue.to_csv('DatasetTagAndValue.csv')
```

### Competitions

```python
competitions = pd.read_csv("Competitions.csv", engine='python')
competitions.head()
```

```python
competitionTags = pd.read_csv("CompetitionTags.csv", engine='python')
competitionTags.head()
```

```python
competitionsAndTags = pd.merge(competitionTags[["CompetitionId", "TagId"]], tagsAndParents, left_on="TagId", right_on="Id")
competitionsAndTags = competitionsAndTags.drop(['Id', 'TagId'], axis=1)

competitionsAndTags = pd.merge(competitions, competitionsAndTags, left_on="Id", right_on="CompetitionId")
competitionsAndTags = competitionsAndTags.drop(['CompetitionId'], axis=1)

competitionsAndTags.head()
```

```python
competitionTagAndValue = competitionsAndTags.groupby(['Name', 'ParentTagName'])['Name'].count()
competitionTagAndValue
```

```python
competitionTagAndValue.to_csv('CompetitionTagAndValue.csv')
```
