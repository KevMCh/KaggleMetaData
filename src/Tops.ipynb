### Dataset Version

```python
datasetVersions = pd.read_csv("DatasetVersions.csv", engine='python', encoding='utf-8', on_bad_lines='skip', index_col=False)
datasetVersions = datasetVersions[datasetVersions.Id.apply(lambda x: x.isnumeric())]
datasetVersions['Id'] = datasetVersions['Id'].astype(float).astype(int)

datasetVersions = datasetVersions[datasetVersions.DatasetId.apply(lambda x: x.isnumeric())]
datasetVersions['DatasetId'] = datasetVersions['DatasetId'].astype(float).astype(int)

datasetVersions
```

```python
datasets = pd.read_csv("Datasets.csv", engine='python', encoding='utf-8', on_bad_lines='skip', index_col=False)

datasets.head()
```

### Filters

##### Values to filter the users:

```python
top = 25
filterMetric = "Degree"
years = range(2018, 2023)
```

### Genereting and saving top users/datasets over time

```python
def getTopUsers(year, nTop, filterMetric, userInfo):
  userInfo = userInfo.sort_values(by=[filterMetric + str(year)], ascending=False)
  topUsers = userInfo.head(nTop)
  
  topUsers = topUsers.rename(columns={"Id": "UserId"})

  return topUsers
```

```python
def getTopDatasetFromUsers(year, datasetVotes, topUsers, datasets, datasetVersions):
  datasetVotes = datasetVotes[(datasetVotes['VoteDate'] < (str(year) + '-01-01'))]

  topDatasets = pd.merge(datasetVotes, topUsers, on="UserId", how="inner")

  topDatasets = pd.merge(topDatasets[["DatasetVersionId"]], datasetVersions, left_on="DatasetVersionId", right_on="Id", how="inner")

  topDatasets = pd.merge(topDatasets[["DatasetId",	"LicenseName",	"VersionNumber",	"Title",	"Slug",	"Subtitle",	"VersionNotes"]], datasets, left_on="DatasetId", right_on="Id")

  topDatasets = topDatasets[["Id", "Title",	"Subtitle", "CreationDate",	"LicenseName",	"VersionNumber",	"VersionNotes",	"TotalViews", "TotalDownloads",	"TotalVotes"]]

  return topDatasets.drop_duplicates()
```

```python
allDatasets = pd.DataFrame()

for year in years:
  topUsers = getTopUsers(year, top, filterMetric, userInfo)
  topDatasets = getTopDatasetFromUsers(year, datasetVotes, topUsers, datasets, datasetVersions)
  
  tmpTopDatasets = topDatasets
  tmpTopDatasets = tmpTopDatasets.rename(columns={"TotalVotes": "TotalVotes" + str(year)})
  allDatasets = allDatasets.append(tmpTopDatasets)

  topUsers.to_csv('Top' + str(top) + 'Users' + str(year) + '.csv', index=False)
  topDatasets.to_csv('TopDatasets' + str(year) + '.csv', index=False)

allDatasets = allDatasets.groupby(["Id", "Title",	"Subtitle", "CreationDate",	"LicenseName",	"VersionNumber",	"VersionNotes",	"TotalViews", "TotalDownloads"])['TotalVotes2018', 'TotalVotes2019',	'TotalVotes2020',	'TotalVotes2021',	'TotalVotes2022'].sum().reset_index()
allDatasets.to_csv('AllDatasetInfo.csv', index=False)
```
